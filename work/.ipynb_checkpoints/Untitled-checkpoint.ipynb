```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
```

```python
def plot_iterations2(n,y1,y2):
    k = [i for i in range(0,n)]
    x_iterations = [sqrt(2**(n-i)) for i in range(0,n)]
    x1= [a*b for a,b in zip(x_iterations,y1)]
    x2= [a*b for a,b in zip(x_iterations,y2)]
    x_med = [(a+b)/2 for a,b in zip(x1,x2)]
    # plot

    plt.rcParams["figure.figsize"] = (15, 7)
    #plt.rcParams["font.family"] = "Times New Roman"
    plt.rcParams["font.size"] = "30"

    fig, ax = plt.subplots(figsize=(8, 8))


    ax.plot(k, x_med, 'k^-')
    ax.fill_between(k, x2, x1, alpha=0.2, color='silver')

    ax.set_xlabel("Number of fixed qubits",fontsize=30)
    ax.set_ylabel("Number of Grover iterations",fontsize=30)

    ax.set(xlim=(0, 5), xticks=np.arange(1, 5))

    plt.show()
```

```python
def plot_iterations(n,y1,y2):
    k = [i for i in range(0,n)]
    x_iterations = [sqrt(2**(n-i)) for i in range(0,n)]
    x1= [a*b for a,b in zip(x_iterations,y1)]
    x2= [a*b for a,b in zip(x_iterations,y2)]
   
    plt.rcParams["font.size"] = "30"

    fig, ax = plt.subplots()


    ax.plot(k, x1, 'k^-', label = "Maximum number of iterations")
    ax.plot(k, x2, 'ko--', label = "Minimum number of iterations")
    ax.fill_between(k, x2, x1, alpha=0.2, color='silver')

    ax.legend()
    ax.grid()

    ax.set_xlabel("Number of fixed qubits,k",fontsize=30)
    ax.set_ylabel("Number of Grover queries, r",fontsize=30)

    ax.set(xlim=(0, n-1), xticks=np.arange(1, 5),
          ylim=(0,sqrt(2**n)),yticks=np.arange(0,sqrt(2**n)))

    plt.show()
```

```python
n=5
def f(x):
    a=1
    b=0.3
    return a*x+b

k = [i for i in range(0,n)]
x = [2**(n-i) * f((n-i)) for i in range(0,n)]
x_iteration= [2**(n-i) for i in range(0,n)]




# plt.rcParams["figure.figsize"] = (15, 7)
#plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = "30"

fig, ax = plt.subplots()

ax.plot(k, x_iteration, 'k^-', label = "One maximum finding iteration")
ax.plot(k, x, 'ko--', label = "All maximum finding iterations")
ax.fill_between(k, x_iteration, x, alpha=0.2, color='silver')

ax.legend()
ax.grid()

ax.set_xlabel("Number of fixed qubits,k",fontsize=30)
ax.set_ylabel("Number of Grover queries, r",fontsize=30)

ax.set(xlim=(0, n-1), xticks=np.arange(1, n))
plt.title("Theoretical complexity reduction")
plt.show()
```
